```python
# def remove_stopwords(text):
#     word_tokens = word_tokenize(text)
#     word_tokens = [w.replace('&', 'and') for w in word_tokens]
#     filtered_words = [w for w in word_tokens if w.lower() not in stop_words]
#     return ' '.join(filtered_words)
def remove_stopwords(text):
    # Tokenize the text
    word_tokens = word_tokenize(text)
    filtered_words = []
    for w in word_tokens:
        if w == '&':
            filtered_words.append('and')
        elif w.lower() == 'tv':
            filtered_words.append('telvision')
        # elif w.lower() == 'mini-series' or w.lower() == 'miniseries':
        #     filtered_words.append('mini series')
        else:
            if w.lower() not in stop_words:
                filtered_words.append(w)
    return ' '.join(filtered_words)

# Apply the 'remove_stopwords' function to the 'text' column
df['text'] = df['text'].apply(remove_stopwords)

# Save the DataFrame back to a CSV file
df.to_csv(output_file, index=False)
```

```python
def tweet_preprocess(tweet):
    tweet = tweet.lower()
    tweet = emoji.replace_emoji(tweet, replace='')
    tweet = re.sub(hashtag, ' ', tweet)
    tweet = re.sub(tag, ' ', tweet)
    tweet = re.sub(links, ' ', tweet)
    # tweet = re.sub(punc, ' ', tweet)
    # tweet = [word for word in tweet.split() if word.lower() not in stop_words]
    # tweet = " ".join(tweet)
    return tweet
```

```python
def extract_name(quote, language='english'):
  tokens = word_tokenize(quote)
  tags = nltk.pos_tag(tokens)
  tree = nltk.ne_chunk(tags)
  names=[]
  for subtree in tree.subtrees():
      if subtree.label() == 'PERSON':
          leave = " ".join(l[0] for l in subtree.leaves())
          names.append(leave)
  return names
```

```python
def get_hosts(tweets):
    common_names = []
    for tweet in tweets:
        if re.findall(r"\s[hH]osted?",tweet):
            common_names.extend(extract_name(tweet))
    top = nltk.FreqDist(common_names).most_common(5)
    return [''.join(top[0][0]), ''.join(top[1][0])]
```

```python
def string_intersection(s1, s2):
    set1 = set(s1.split())
    set2 = set(s2.split())

    common_words = set1 & set2
    return ' '.join(common_words)

# str1 = "winner jodie foster celebrates 47 years"
# str2 = "jodie foster accepting the cecil b. demille award"

# print(string_intersection(str1, str2))
```

```python
def longest_common_substring(s1, s2):
    m, n = len(s1), len(s2)
    # Create a table to store lengths of the longest common suffixes of substrings.
    dp = [[0] * (n+1) for _ in range(m+1)]
    length = 0  # To store the length of the longest common substring.
    end = 0  # To store the ending index of the longest common substring in s1.

    # Build the dp table in a bottom-up manner.
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1] or s1[i-1] == ' ' or s2[j-1] == ' ':  # Ignore spaces for comparison
                dp[i][j] = dp[i-1][j-1] + 1
                if dp[i][j] > length:
                    length = dp[i][j]
                    end = i
            else:
                dp[i][j] = 0

    # Return the longest common substring.
    return s1[end-length:end]
```
